## Introduction

### Importing Required Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import the spreadsheet file and assign it to the variable df
df = pd.read_csv('diabetes.csv')
```

```python
# Display the first 10 rows of the dataframe
(df.head(10))
```

### Data Verification

```python
# Display the columns in the data file
(df.columns)
```

### Relational Plots

```python
sns.scatterplot(
    data=df,
    x='Age',
    y='BloodPressure'
).set(
    title='Age and Blood Pressure',
    xlabel='Age [years]',
    ylabel='Blood Pressure [mmHg]'
)
```

### Distribution Plots

```python
sns.displot(
    df,
    x='Age',
    y='BloodPressure'
).set(
    title='Ages and Blood Pressure',
    xlabel='Age [years]',
    ylabel='Blood Pressure [mmHg]'
)
```

### Bar Plot

```python
# Create a bar graph to show the difference between Age in years vs Blood Pressure in mmHg and also tilt the x-axis labels by 45 degrees
sns.barplot(
    data=df,
    x='Age',
    y='BloodPressure'
).set(
    title='Age vs Blood Pressure',
    xlabel='Age [years]',
    ylabel='Blood Pressure [mmHg]'
)
# Rotate the x-axis labels by 45 degrees 
plt.xticks(rotation=45)

# enlarge the figure size
plt.figure(figsize=(50,30))
```
